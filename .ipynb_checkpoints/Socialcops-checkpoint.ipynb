```python
def get_subset_by_IQR(df,column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    mask = (df[column] >= (q1-1.5*iqr)) & (df[column] <= (q3+1.5*iqr))
    return df.loc[mask]
```

```python
for i in monthly_data['APMC'].unique():
    for j in monthly_data.loc[monthly_data['APMC'] == i , 'commodity'].unique():
        
        monthly_data.loc[(monthly_data['APMC']==i) & (monthly_data['commodity']==j)] = (get_subset_by_IQR(monthly_data.loc[(monthly_data['APMC'] == i) & (monthly_data['commodity'] ==j) ] , 'modal_price'))
```

```python
monthly_data.dropna(axis=0 , inplace=True)
```

```python

def season_type_apmc(df , i):
    result1 = sm.tsa.seasonal_decompose(df[df['commodity'] == i]['modal_price'].values , freq=12 , model='additive')
    result2 = sm.tsa.seasonal_decompose(df[df['commodity'] == i]['modal_price'].values , freq=12 , model='multiplicative')
    
    sse1 = np.sum( (x-result1.resid.mean())**2  for x in result1.resid)
    sse2 = np.sum( (x-result2.resid.mean())**2  for x in result2.resid )
    
    if(sse1 < sse2):
        category = 'additive'
    else:
        category = 'multiplicative'
        
    return category

for i in index_date['commodity'].unique():
    print(season_type_apmc(index_date , i))
```
